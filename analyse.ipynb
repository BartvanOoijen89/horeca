```python
import pandas as pd
import requests
import zipfile
import tempfile

def download_knmi_daily(station="370", startjaar=2024, eindjaar=2025):
    BASE_URL = "https://cdn.knmi.nl/knmi/map/page/klimatologie/gegevens/daggegevens/"
    url = f"{BASE_URL}dag{station}{startjaar}{eindjaar}.zip"
    r = requests.get(url)
    if r.status_code != 200:
        print("Data niet gevonden of verkeerd jaar/station. Gebruik handmatig download via KNMI website.")
        return None
    # Save zip to temp file, extract and read CSV
    tmp = tempfile.NamedTemporaryFile(delete=False)
    tmp.write(r.content)
    tmp.close()
    with zipfile.ZipFile(tmp.name, 'r') as zip_ref:
        txt_filename = zip_ref.namelist()[0]
        with zip_ref.open(txt_filename) as txtfile:
            df = pd.read_csv(txtfile, skiprows=31, sep=",")
    return df

# Download data van Volkel voor de jaren 2024-2025 (pas aan indien nodig)
knmi_df = download_knmi_daily(station="370", startjaar=2024, eindjaar=2025)

# Verwerk de data
if knmi_df is not None:
    knmi_df['YYYYMMDD'] = pd.to_datetime(knmi_df['YYYYMMDD'], format='%Y%m%d')
    knmi_df['Temp'] = knmi_df['TG'] / 10  # Temperatuur in °C
    knmi_df['Neerslag'] = knmi_df['RH'] / 10  # Neerslag in mm
    weerdata_knmi = knmi_df[['YYYYMMDD', 'Temp', 'Neerslag']].rename(columns={'YYYYMMDD': 'Datum'})
    print(weerdata_knmi.head())
else:
    print("KNMI-data niet gedownload.")
```

```python
import pandas as pd
import glob
import os
import re

# 1. Zoek alle verkoopbestanden in de map 'verkopen'
verkoop_bestanden = glob.glob(os.path.join('verkopen', '*.csv'))

verkoop_df_lijst = []

for bestand in verkoop_bestanden:
    print(f"Inlezen: {bestand}")  # Debug, mag je weghalen
    
    # Locatie en datum uit bestandsnaam halen (voorbeeld: Verkochte-Producten-Oranjerie_09-06-2025.csv)
    filename = os.path.basename(bestand)
    match = re.match(r'Verkochte-Producten-(.*?)_(\d{2}-\d{2}-\d{4})\.csv', filename)
    if match:
        locatie = match.group(1)
        datum_str = match.group(2)
        datum = pd.to_datetime(datum_str, dayfirst=True)
    else:
        locatie = "Onbekend"
        datum = pd.NaT

    try:
        # Laat pandas het juiste scheidingsteken raden (werkt voor komma én puntkomma)
        temp_df = pd.read_csv(bestand, sep=None, engine='python')
    except Exception as e:
        print(f"FOUT in bestand {bestand}: {e}")
        continue

    # Voeg locatie en datum toe aan de dataframe
    temp_df['locatie'] = locatie
    temp_df['datum'] = datum

    verkoop_df_lijst.append(temp_df)

# 2. Voeg alles samen tot één dataframe
verkoop_df = pd.concat(verkoop_df_lijst, ignore_index=True)

# 3. Aggregeren per dag en per locatie
verkoop_per_dag = (
    verkoop_df
    .groupby(['datum', 'locatie'])
    .agg({'Netto omzet incl. btw': 'sum', 'Aantal': 'sum'})
    .reset_index()
)

print(verkoop_per_dag.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# == INPUTS ==
DATUM = pd.to_datetime("2025-07-21")
BEGROOT = 200
TEMP = 19.0
NEERSLAG = 0.0
VOORSPELLEN = "aantal"  # of 'omzet'

# == DATA (neem hier je gemergede dataframe!) ==
# Bijvoorbeeld df_merged = ...
df = df_merged.copy()

# Lijsten voor alle unieke locaties en omzetgroepen
ALLE_LOCATIES = df['locatie'].dropna().unique()
ALLE_OMZETGROEPEN = df['omzetgroep naam'].dropna().unique()

# == Functie: voorspellen per product ==
def voorspel_per_product(df, productnaam, locatie, omzetgroep, datum, begroot, temp, neerslag, target='aantal'):
    df_p = df[
        (df['product name'] == productnaam) &
        (df['locatie'] == locatie) &
        (df['omzetgroep naam'] == omzetgroep)
    ]
    # target: aantal of omzet
    if target == 'omzet':
        y_col = 'netto omzet incl. btw'
    else:
        y_col = 'aantal'
    # Drop NaN rows
    df_p = df_p.dropna(subset=[y_col, 'begroot aantal bezoekers', 'Temp', 'Neerslag'])
    # Als te weinig data: fallback naar omzetgroep-gemiddelde (optioneel!)
    status = "ok"
    if len(df_p) < 3:
        # Fallback: omzetgroep-niveau
        df_group = df[
            (df['locatie'] == locatie) & 
            (df['omzetgroep naam'] == omzetgroep)
        ].dropna(subset=[y_col, 'begroot aantal bezoekers', 'Temp', 'Neerslag'])
        if len(df_group) >= 3:
            X = df_group[['begroot aantal bezoekers', 'Temp', 'Neerslag']]
            y = df_group[y_col]
            model = LinearRegression().fit(X, y)
            voorspeld = model.predict([[begroot, temp, neerslag]])[0]
            status = "groep-voorspelling"
        else:
            voorspeld = 0
            status = "onvoldoende data"
    else:
        X = df_p[['begroot aantal bezoekers', 'Temp', 'Neerslag']]
        y = df_p[y_col]
        model = LinearRegression().fit(X, y)
        voorspeld = model.predict([[begroot, temp, neerslag]])[0]
    # Geen negatieve aantallen/omzet
    voorspeld = max(0, round(voorspeld))
    return voorspeld, status

# == OVERZICHT: alle locaties, omzetgroepen, producten ==
resultaat = []
for locatie in ALLE_LOCATIES:
    for omzetgroep in ALLE_OMZETGROEPEN:
        producten = df[(df['locatie'] == locatie) & (df['omzetgroep naam'] == omzetgroep)]['product name'].dropna().unique()
        for p in producten:
            verwacht, status = voorspel_per_product(
                df, p, locatie, omzetgroep, DATUM, BEGROOT, TEMP, NEERSLAG, VOORSPELLEN
            )
            resultaat.append({
                'datum': DATUM.date(),
                'locatie': locatie,
                'omzetgroep': omzetgroep,
                'product name': p,
                'verwacht {}'.format(VOORSPELLEN): verwacht,
                'status': status
            })

# Alles naar DataFrame
df_resultaat = pd.DataFrame(resultaat)

# == PRINT: per locatie & omzetgroep ==
for locatie in ALLE_LOCATIES:
    print(f"\n=== {VOORSPELLEN.title()}-voorspelling voor {locatie} op {DATUM.date()} ===")
    df_loc = df_resultaat[df_resultaat['locatie'] == locatie]
    if df_loc.empty:
        print("Geen producten/omzetgroepen voor deze locatie.")
    else:
        for omzetgroep in ALLE_OMZETGROEPEN:
            df_groep = df_loc[df_loc['omzetgroep'] == omzetgroep]
            if not df_groep.empty:
                print(f"\n-- {omzetgroep} --")
                print(df_groep[['product name', f"verwacht {VOORSPELLEN}", 'status']].to_string(index=False))
                print("Totaal {}: {}".format(
                    VOORSPELLEN,
                    df_groep[f"verwacht {VOORSPELLEN}"].sum()
                ))
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import glob
import re

### === INLEZEN EN VOORBEREIDEN DATA ===

# 1. Verkoopbestanden inlezen
bestanden = glob.glob('verkopen/Verkochte-Producten-*.csv')
dfs = []
for f in bestanden:
    match = re.search(r'Verkochte-Producten-(.*)_(\d{2}-\d{2}-\d{4})\.csv', f)
    if match:
        locatie = match.group(1)
        datum = pd.to_datetime(match.group(2), format='%d-%m-%Y')
    else:
        locatie = 'Onbekend'
        datum = pd.NaT
    df_temp = pd.read_csv(f, sep=';', decimal=',')
    df_temp['locatie'] = locatie
    df_temp['datum'] = datum
    df_temp.columns = df_temp.columns.str.strip().str.lower()
    dfs.append(df_temp)
df = pd.concat(dfs, ignore_index=True)
for col in ['aantal', 'netto omzet incl. btw']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df['aantal'] = df['aantal'].fillna(0).astype(int)

# 2. Agregeren op dag/locatie/omzetgroep/product
df_aggr = (
    df.groupby(['datum', 'locatie', 'omzetgroep naam', 'product name'])
    .agg({'aantal': 'sum', 'netto omzet incl. btw': 'sum'})
    .reset_index()
)

ALLE_LOCATIES = df['locatie'].unique()
ALLE_OMZETGROEPEN = df['omzetgroep naam'].unique()
ALLE_PRODUCTEN = df['product name'].unique()

# 3. Bezoekersdata
bezoekers_df = pd.read_excel('Bezoekersdata.xlsx')
bezoekers_df.columns = bezoekers_df.columns.str.strip().str.lower()
bezoekers_df['datum'] = pd.to_datetime(bezoekers_df['datum'])
if 'locatie' not in bezoekers_df.columns:
    bezoekers_df['locatie'] = 'Totaal'

# 4. Weerdata (KNMI)
weerdata_knmi = pd.read_csv(
    'Volkel_weerdata.txt',
    skiprows=47, usecols=[1, 11, 21], names=['YYYYMMDD', 'TG', 'RH']
)
weerdata_knmi['YYYYMMDD'] = weerdata_knmi['YYYYMMDD'].astype(str)
weerdata_knmi['datum'] = pd.to_datetime(weerdata_knmi['YYYYMMDD'], format='%Y%m%d', errors='coerce')
weerdata_knmi['TG'] = pd.to_numeric(weerdata_knmi['TG'], errors='coerce')
weerdata_knmi['RH'] = pd.to_numeric(weerdata_knmi['RH'], errors='coerce')
weerdata_knmi['Temp'] = weerdata_knmi['TG'] / 10
weerdata_knmi['Neerslag'] = weerdata_knmi['RH'].clip(lower=0) / 10
weerdata = weerdata_knmi[['datum', 'Temp', 'Neerslag']].copy()

# 5. Koppelen bezoekers/weerdata aan verkoopdata (altijd park-totaal bezoekers)
bezoekers_park = bezoekers_df.groupby('datum').agg({
    'begroot aantal bezoekers': 'sum',
    'totaal aantal bezoekers': 'sum'
}).reset_index()
df_aggr = pd.merge(df_aggr, bezoekers_park, on='datum', how='left')
df_aggr = pd.merge(df_aggr, weerdata, on='datum', how='left')

### === VOORSPELLINGSFUNCTIES ===

def voorspel_per_locatie(
    locatie, datum, begroot, temp, neerslag, min_dagen=3, tonen_omzet=True
):
    print(f"\n=== Voorspelling voor {locatie} op {datum.date()} ===\n")
    totaal_omzet = 0
    iets_voorspeld = False

    for omzetgroep in ALLE_OMZETGROEPEN:
        # Selecteer alle producten voor deze locatie+omzetgroep
        producten = df_aggr.loc[
            (df_aggr['locatie'] == locatie) & (df_aggr['omzetgroep naam'] == omzetgroep),
            'product name'
        ].unique()
        resultaten = []

        for prod in producten:
            df_p = df_aggr[
                (df_aggr['locatie'] == locatie) &
                (df_aggr['omzetgroep naam'] == omzetgroep) &
                (df_aggr['product name'] == prod)
            ]
            df_p = df_p.dropna(subset=['begroot aantal bezoekers', 'Temp', 'Neerslag', 'aantal', 'netto omzet incl. btw'])
            if len(df_p) < min_dagen:
                continue  # skip als te weinig trainingsdata

            X = df_p[['begroot aantal bezoekers', 'Temp', 'Neerslag']]
            y = df_p['aantal']
            y_omzet = df_p['netto omzet incl. btw']
            model = LinearRegression().fit(X, y)
            model_omzet = LinearRegression().fit(X, y_omzet)

            x_voorspel = pd.DataFrame({
                'begroot aantal bezoekers': [begroot],
                'Temp': [temp],
                'Neerslag': [neerslag]
            })

            aantal = int(round(model.predict(x_voorspel)[0]))
            aantal = max(0, aantal)
            if aantal == 0:
                continue  # Niet tonen als de voorspelling 0 is

            iets_voorspeld = True
            resultaten.append({'product name': prod, 'verwacht aantal': aantal, 'status': 'voorspeld'})

            # Omzet voor dit product erbij optellen (voor totaal per locatie)
            if tonen_omzet:
                omzet = model_omzet.predict(x_voorspel)[0]
                omzet = max(0, omzet)
                totaal_omzet += omzet

        if resultaten:
            print(f"-- {omzetgroep} --")
            df_res = pd.DataFrame(resultaten)
            print(df_res[['product name', 'verwacht aantal', 'status']].to_string(index=False))
            print(f"Totaal aantal: {df_res['verwacht aantal'].sum()}\n")

    if tonen_omzet and totaal_omzet > 0:
        print(f">>> Verwachte omzet totaal voor {locatie}: € {totaal_omzet:,.2f}\n")
    elif tonen_omzet:
        print(f">>> Geen omzetvoorspelling voor {locatie} (onvoldoende data)\n")
    if not iets_voorspeld:
        print("-- Geen voorspellingen voor deze dag / locatie --\n")

### === GEBRUIK: invoer voor voorspelling ===

# Stel de voorspeldata in:
DATUM = pd.Timestamp("2025-07-21")
bezoekers_op_dag = bezoekers_park[bezoekers_park['datum'] == DATUM]
if len(bezoekers_op_dag) == 0:
    begroot = 0
else:
    begroot = int(bezoekers_op_dag['begroot aantal bezoekers'].values[0])
temp = weerdata.loc[weerdata['datum'] == DATUM, 'Temp'].values[0] if DATUM in weerdata['datum'].values else 20
neerslag = weerdata.loc[weerdata['datum'] == DATUM, 'Neerslag'].values[0] if DATUM in weerdata['datum'].values else 0.0

# Draai voorspelling per locatie
for locatie in ALLE_LOCATIES:
    voorspel_per_locatie(locatie, DATUM, begroot, temp, neerslag, min_dagen=3, tonen_omzet=True)
```
